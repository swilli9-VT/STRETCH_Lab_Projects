```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib import cm
# import itertools as product
```

```python
plt.rcParams.update({'font.size': 16})
plt.rcParams['axes.linewidth'] = 1.5

### Initialize dataframe for results storage
R = pd.read_csv("C:/Users/wdsnyde/code/fhnm-ldrd/Docs/Schwarz-PINNs/OverlapResults.csv")
#R = pd.read_csv("C:/Users/wdsnyde/code/fhnm-ldrd/Docs/Schwarz-PINNs/ParamResults_FOM_None.csv")

n_sub = R["Sub-domains"].unique()
overlap = R["% Overlap"].unique()*100

X, Y = np.meshgrid(n_sub,overlap)

Z = np.zeros(X.shape)

array1 = R.values[:,1:]

for k in [1]:#range(1,7):
    Pe = 10**k
    tempPe = array1[array1[:,0]==Pe]
    
    fig, ax = plt.subplots(2,2, figsize=(15,10))#, subplot_kw={"projection": "3d"})
    fig.tight_layout(pad=5.0)
    fig.suptitle("Pe = "+str(Pe), fontsize=32, x = 0.03, y=0.6, rotation=90)
    
    l = 0
    for s in (False, True):
        tempSnap = tempPe[tempPe[:,2]==s]
        slabel = str(s)
        
        w = 0
        for b in ("weak", "strong"):
            tempBC = tempSnap[tempSnap[:,3]==b]
            temp = tempBC[:,4:]
            
            label = "Snapshots: "+slabel+"; BC Enforcement: "+b
            
            c = 0  
            for j in range(X.shape[1]):
                for i in range(X.shape[0]):
                    #check = temp[c][4:]
                    #val = check[np.where(~pd.isnull(check))]
                    #Z[i][j] = np.sum(val)/len(val)
                    
                    #Z[i][j] = temp[c][2]
                    Z[i][j] = temp[c][4]
                    
                    c += 1
            
            vmin = 0
            vmax = 9e-3
            ax[l, w].set_xlabel('Sub-domains', fontsize=16)
            ax[l, w].set_xticks(n_sub)
            ax[l, w].set_ylabel('% Overlap', fontsize=16)
            #ax[l, w].set_yticks(overlap)
            #ax[l, w].set_zlabel('CPU Time (s)')
            ax[l, w].set_title(label, fontsize=16)
            surf = ax[l, w].contourf(X, Y, Z, antialiased=False, cmap=cm.coolwarm, vmin=vmin, vmax=vmax)
            m = plt.cm.ScalarMappable(cmap=cm.get_cmap('coolwarm', 7))
            m.set_clim(vmin, vmax)
#             surf = ax[l, w].plot_surface(X, Y, Z, cmap=cm.coolwarm,
#                        linewidth=0, antialiased=False)
            
            cb = fig.colorbar(m, ax = ax[l, w], ticks=np.linspace(vmin,vmax, num=8),
                 shrink = 0.5, aspect = 5, pad = 0.04, format="%1.1e")
            
            cb.set_label(label = 'Avg L2 Error Across Sub-domains',size=16,rotation=90,verticalalignment='top')
            w+=1
        l+=1
    fig.savefig("C:/Users/wdsnyde/code/fhnm-ldrd/Docs/Schwarz-PINNs/PINN-PINN_Parameter_Sweep/AvgL2_sweep_Pe_{:d}.eps".format(Pe), dpi=3000)
```
